```python
%load_ext autoreload
%autoreload 2

# Import libaries
import pandas as pd

import sys
sys.path.insert(0, '../')
import src.cleanup as cleanup
```

```python
# Import csv as churners
churners = pd.read_csv('../data/BankChurners.csv')
churners.head()
```

```python
# Drop unnecessary coulmns from csv
churners = cleanup.drop_column(churners, ['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2' ])
```

```python
# Check for null data
print(churners.isnull().sum())
```

```python
# Get dummies for target column
# column_name must be a list
churners = cleanup.dummies(churners, 'Attrition_Flag')
churners.head()
```

```python
# Move encoded columns
churners = cleanup.move_column(churners, 'Attrition_Flag_Attrited Customer', 1)
display(churners)
```

```python
# Rename attrited column
churners = cleanup.column_rename(churners, 'Attrition_Flag_Attrited Customer','Attrited')
display(churners)
```

```python
# Get dummies for gender column
churners = cleanup.dummies(churners, 'Gender')
display(churners)
```

```python
# Get dummies for marital status
churners = cleanup.dummies(churners, 'Marital_Status')
display(churners)
```

```python
# Encode 'Education Level'
education_hierarchy = ['Unknown', 'Uneducated', 'High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate']
churners = cleanup.encode_ordinal(churners, 'Education_Level', education_hierarchy)
display(churners)
```

```python
# Encode Income_Category
income_hierarchy = ['Unknown', 'Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +']
churners = cleanup.encode_ordinal(churners, 'Income_Category', income_hierarchy)
display(churners)
```

```python
# Encode Card_Category
card_hierarchy = ['Blue', 'Silver', 'Gold', 'Platinum']
churners = cleanup.encode_ordinal(churners, 'Card_Category', card_hierarchy)
display(churners)
```

```python
# Change booleans to binary
churners = cleanup.replace_boolean(churners)
churners.head()
```

```python
# Columns are dropped to avoid collinearity

churners = cleanup.drop_column(churners, 'Gender_F')
churners = cleanup.drop_column(churners, 'Marital_Status_Unknown')
churners = cleanup.drop_column(churners, 'Attrition_Flag_Existing Customer')
churners.head()
```
